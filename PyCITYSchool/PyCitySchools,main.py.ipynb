```python
import pandas as pd
import numpy as np
```

```python
school_data = pd.read_csv(r"C:\Users\nicho\OneDrive\Desktop\Starter_Code (3)\Starter_Code\PyCitySchools\Resources\schools_complete.csv")
```

```python
student_data = pd.read_csv(r"C:\Users\nicho\OneDrive\Desktop\Starter_Code (3)\Starter_Code\PyCitySchools\Resources\students_complete.csv")
```

```python
merge_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
merge_df.head()
```

```python
school_names = merge_df['school_name'].unique()
school_names
total_schools = len(school_names)
total_schools
```

```python
students = merge_df['student_name']
total_students = len(students)
total_students
```

```python
budget = merge_df['budget']
total_budget = sum(budget)
total_budget
```

```python
math = merge_df['math_score']
avg_mscore = math.mean()
avg_mscore
```

```python
reading = merge_df['reading_score']
avg_rscore = reading.mean()
avg_rscore
```

```python
math_passcount = merge_df[(merge_df['math_score'] >= 70)].count()['student_name']
math_passperc = math_passcount / float(total_students) * 100
math_passperc
```

```python
read_passcount = merge_df[(merge_df['reading_score'] >= 70)].count()['student_name']
read_passperc = read_passcount / float(total_students) *100
read_passperc
```

```python
math_read_passcount = merge_df[
    (merge_df["math_score"] >= 70) & (merge_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = math_read_passcount / (total_students) * 100
overall_passing_rate
```

```python
district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_mscore],
    "Average Reading Score": [avg_rscore],
    "% Passing Math": [math_passperc],
    "% Passing Reading": [read_passperc],
    "% Overall Passing Rate": [overall_passing_rate]
})

district_summary_df
```
